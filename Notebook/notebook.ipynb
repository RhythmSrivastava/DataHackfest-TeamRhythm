```python
import pandas as pd
import os
import glob

# Define the base directory where the subdirectories are stored
base_directory = r"C:\Users\mailg\OneDrive\Documents\GitHub\DataHackfest-TeamRhythm\Dataset\almrrc2021_data"

# Use glob to find all JSON files in subdirectories
json_files = glob.glob(os.path.join(base_directory, "**", "*.json"), recursive=True)

# Create a dictionary to store dataframes with filenames as keys
dataframes = {}

# Load the JSON files into separate dataframes
for file in json_files:
    # Extract the file name without extension to use as dataframe name
    file_name = os.path.splitext(os.path.basename(file))[0]
    # Load the dataframe
    dataframes[file_name] = pd.read_json(file)
    # Display basic information and the first few rows of each dataframe
    print(f"DataFrame '{file_name}':")
    print(dataframes[file_name].info())
    print(dataframes[file_name].head(), "\n")

# The dataframes dictionary now contains each dataframe with the corresponding file name
```
